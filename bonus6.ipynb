```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('imdb.csv',escapechar='\\')
```

```python
df.head()
```

```python
df1 = df.groupby('year')['imdbRating'].mean().reset_index()
```

```python
df1 = pd.DataFrame(df1)
df1
```

```python
plt.plot(df1['year'],df1['imdbRating'])
plt.xlabel('Year')
plt.ylabel('imdbRating')
plt.title('imdb Rating over years')
```

```python
sns.barplot(x='type',y='duration',data=df)
```

```python
sns.countplot(x='type',data=df)
```

```python
df2=pd.DataFrame(df["type"].str.split(".", n = 1, expand = True))
df2.columns=['count','type']
df2=df2.groupby('type').count().reset_index()
plt.pie(df2['count'], labels=df2['type'],autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Types of videos')
plt.axis('equal')
plt.show()
```

```python
df.dropna(how='any',axis = 0, inplace=True)
```

```python
sns.distplot(df['year'],kde=False)
```

```python
sns.boxplot(x='type',y='imdbRating',data=df)
```
